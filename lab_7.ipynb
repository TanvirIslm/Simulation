```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Part 1: Generate Random Sample
# -----------------------------
mean = 100
std = 20
sample_size = 200

sample = np.random.normal(mean, std, sample_size)

print("Sample Size =", sample_size)
print("Sample Mean =", round(np.mean(sample), 2))
print("Sample Standard Deviation =", round(np.std(sample), 2))

# -----------------------------
# Part 2: Histogram (Normal Distribution)
# -----------------------------
plt.figure(figsize=(8,5))
plt.hist(sample, bins=15, density=True)

plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Histogram of Normal Distribution (μ=100, σ=20)")
plt.grid(True)
plt.show()

# -----------------------------
# Part 3: Unimodal Density Curve
# -----------------------------
x = np.linspace(0, 200, 1000)

mu = 100
sigma = 20

y = (1/(sigma*np.sqrt(2*np.pi))) * \
    np.exp(-((x-mu)**2)/(2*sigma**2))

plt.figure(figsize=(8,5))
plt.plot(x, y)

plt.title("Unimodal Normal Density Curve")
plt.xlabel("X")
plt.ylabel("Density")
plt.grid(True)
plt.show()

# -----------------------------
# Part 4: Multimodal Density Curve
# -----------------------------
mu1 = 70
sigma1 = 15

mu2 = 130
sigma2 = 15

y1 = (1/(sigma1*np.sqrt(2*np.pi))) * \
     np.exp(-((x-mu1)**2)/(2*sigma1**2))

y2 = (1/(sigma2*np.sqrt(2*np.pi))) * \
     np.exp(-((x-mu2)**2)/(2*sigma2**2))

multi = y1 + y2

plt.figure(figsize=(8,5))
plt.plot(x, multi)

plt.title("Multimodal Density Curve")
plt.xlabel("X")
plt.ylabel("Density")
plt.grid(True)
plt.show()

# -----------------------------
# Part 5: Blood Pressure Example
# -----------------------------
bp_mean = 80
bp_std = 20

blood_pressure = np.random.normal(bp_mean, bp_std, 1000)

plt.figure(figsize=(8,5))
plt.hist(blood_pressure, bins=20, density=True)

plt.xlabel("Diastolic Blood Pressure")
plt.ylabel("Density")
plt.title("Distribution of Diastolic Blood Pressure")
plt.grid(True)
plt.show()
```

```python
#Task 1
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mean = 100
std = 20

x = np.linspace(20,180,500)

y = norm.pdf(x, mean, std)

plt.figure(figsize=(8,5))

plt.plot(x, y)

plt.title("Unimodal Normal Distribution")

plt.xlabel("Value")

plt.ylabel("Density")

plt.grid(True)

plt.show()
```

```python
#Task 2
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.linspace(20,220,600)

y1 = norm.pdf(x,80,10)

y2 = norm.pdf(x,120,15)

y3 = norm.pdf(x,170,12)

plt.figure(figsize=(8,5))

plt.plot(x,y1,label="Mean=80")

plt.plot(x,y2,label="Mean=120")

plt.plot(x,y3,label="Mean=170")

plt.title("Multimodal Density Curves")

plt.xlabel("Value")

plt.ylabel("Density")

plt.legend()

plt.grid(True)

plt.show()
```
