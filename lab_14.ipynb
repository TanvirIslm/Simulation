```python
import random

# Parameters
NUM_CUSTOMERS = 10

# Initial State
clock = 0.0
server_busy = False
queue = []

next_arrival = 1.0
next_departure = float('inf')

customer_id = 1

print("TRACE OF DISCRETE EVENT SIMULATION")
print("-" * 70)
print("Time\tEvent\t\tQueue Length\tServer Status")

while customer_id <= NUM_CUSTOMERS or server_busy:

    event_time = min(next_arrival, next_departure)
    clock = event_time

    # Arrival Event
    if next_arrival <= next_departure and customer_id <= NUM_CUSTOMERS:

        print(f"{clock:.2f}\tArrival C{customer_id}\t{len(queue)}\t\t{'Busy' if server_busy else 'Idle'}")

        if not server_busy:
            server_busy = True
            service_time = random.randint(2, 4)
            next_departure = clock + service_time
        else:
            queue.append(customer_id)

        customer_id += 1

        if customer_id <= NUM_CUSTOMERS:
            next_arrival = clock + random.randint(1, 3)
        else:
            next_arrival = float('inf')

    # Departure Event
    else:

        print(f"{clock:.2f}\tDeparture\t{len(queue)}\t\tBusy")

        if queue:
            queue.pop(0)
            service_time = random.randint(2, 4)
            next_departure = clock + service_time
        else:
            server_busy = False
            next_departure = float('inf')

print("-" * 70)
print("Simulation Finished")
print("Final Clock =", round(clock, 2))
```
