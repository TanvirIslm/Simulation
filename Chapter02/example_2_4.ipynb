```python
import random
import matplotlib.pyplot as plt

N = 5000
M = 0  

x_in, y_in = [], []
x_out, y_out = [], []

for i in range(N):
    x = random.uniform(0, 1)
    y = random.uniform(0, 1)

    if x*x + y*y <= 1:
        M += 1
        x_in.append(x)
        y_in.append(y)
    else:
        x_out.append(x)
        y_out.append(y)

pi_estimate = 4 * M / N

print("Total points (N):", N)
print("Points inside circle (M):", M)
print("Estimated value of pi:", pi_estimate)

plt.figure()
plt.scatter(x_in, y_in, s=5, label="Inside circle")
plt.scatter(x_out, y_out, s=5, label="Outside circle")

theta = [i * 0.01 for i in range(0, 158)]
xs = [x for x in theta]
ys = [(1 - x*x)**0.5 for x in xs]
plt.plot(xs, ys, linewidth=2, label="x² + y² = 1")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Monte Carlo Simulation to Estimate π")
plt.legend()
plt.axis("equal")
plt.show()
```
