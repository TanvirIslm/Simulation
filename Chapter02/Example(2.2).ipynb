Example 2.2. A Gambling Game

To illustrate the case of a random process, let us consider a game in which an unbiased coin is repeatedly flipped. For each flip you have to pay Re. 1. and when the difference between heads tossed
and tails tossed becomes 3, you get Rs. 8. Thus if the required difference is obtained in less than 8 flips, you win some money and if it is in more than 8 flips you lose. 

```python
import random

def run_simulation(num_games):

    print("Game Sl  Rand  H/T   Heads Tails Diff")
    print("-------------------------------------")

    for game in range(1, num_games + 1):
        heads = tails = 0
        sl = 1

        while abs(heads - tails) < 3:
            r = random.randint(0, 9)

            if r <= 4:
                result = 'H'
                heads += 1
            else:
                result = 'T'
                tails += 1

            diff = abs(heads - tails)
            game_no = game if sl == 1 else ""

            print(f"{game_no:<4} {sl:<3} {r:<5} {result:<4} {heads:<5} {tails:<5} {diff}")

            sl += 1

        flips = sl - 1
        profit = 8 - flips

        if profit > 0:
            print(f"{'':<25}Win Re. {profit}")
        else:
            print(f"{'':<25}Lose Re. {abs(profit)}")

        print("-------------------------------------")

run_simulation(2)
```

```python
import random

def simulate_gambling_game_detailed(num_games):
    total_flips = 0
    total_cost = 0
    total_winnings = 0

    print(f"{'Game':<6} {'Sl.':<4} {'Rand':<6} {'H/T':<4} {'Cum H':<6} {'Cum T':<6} {'Diff'}")
    print("-" * 45)

    for game_num in range(1, num_games + 1):
        heads = 0
        tails = 0
        sl_no = 1
        
        while abs(heads - tails) < 3:
            # Generate random number 0-9
            rand_num = random.randint(0, 9)
            
            # 0-4 = Head, 5-9 = Tail
            if rand_num <= 4:
                result = 'H'
                heads += 1
            else:
                result = 'T'
                tails += 1
            
            diff = abs(heads - tails)
            
            # Print row for the current flip
            game_label = game_num if sl_no == 1 else ""
            print(f"{game_label:<6} {sl_no:<4} {rand_num:<6} {result:<4} {heads:<6} {tails:<6} {diff}")
            
            sl_no += 1
            total_cost += 1

        # Calculate result for this game
        # Win if flips < 8 (Cost < 8), Lose if flips > 8 (Cost > 8)
        game_flips = sl_no - 1
        total_winnings += 8
        net = 8 - game_flips
        
        if net > 0:
            print(f"{'':<32} Win Re. {net}")
        elif net < 0:
            print(f"{'':<32} Lose Re. {abs(net)}")
        else:
            print(f"{'':<32} Break Even")
        print("-" * 45)

    avg_flips = total_cost / num_games
    print(f"\nSummary of {num_games} games:")
    print(f"Total Flips (Cost): Rs. {total_cost}")
    print(f"Total Returns: Rs. {total_winnings}")
    print(f"Average flips per game: {avg_flips:.2f}")
    print(f"Net Profit/Loss: Rs. {total_winnings - total_cost}")

# Run for 2 games to match your image format
simulate_gambling_game_detailed(2)
```
