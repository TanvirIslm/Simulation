```python
import random

def simulate_random_walk(steps):
    
    x, y = 0, 0
    
    
    print(f"{'Step':<6} {'Random No.':<12} {'Direction':<12} {'Position (x, y)'}")
    print("-" * 50)
    
    for step in range(1, steps + 1):
        
        rand_num = random.randint(0, 9)
        
    
        if rand_num <= 4:  # 0, 1, 2, 3, 4
            direction = "F (Forward)"
            y += 1
        elif rand_num <= 7:  # 5, 6, 7
            direction = "L (Left)"
            x -= 1
        else:  
            direction = "R (Right)"
            x += 1
            
        
        print(f"{step:<6} {rand_num:<12} {direction:<12} ({x}, {y})")

    print("-" * 50)
    print(f"Final Position after {steps} steps: ({x}, {y})")

simulate_random_walk(50)
```

Forward (F),+y,0.7,30%,00−29
Backward (B),−y,0.6,20%,30−49
Left (L),−x,0.3,15%,50−64
Right (R),+x,0.4,15%,65−79
Up (U),+z,0.2,10%,80−89
Down (D),−z,0.3,10%,90−99

```python
import random

def simulate_3d_random_walk(steps):
    # Starting position in 3D space
    x, y, z = 0.0, 0.0, 0.0
    
    print(f"{'Step':<5} {'RN':<4} {'Dir':<12} {'Step Size':<10} {'Position (x, y, z)':<20}")
    print("-" * 65)
    
    for step in range(1, steps + 1):
        # Generate a 2-digit random number (0-99)
        rn = random.randint(0, 99)
        
        # Logic based on the probability table
        if rn < 30:  # 30% Forward
            direction = "Forward"
            move_size = 0.7
            y += move_size
        elif rn < 50:  # 20% Backward
            direction = "Backward"
            move_size = 0.6
            y -= move_size
        elif rn < 65:  # 15% Left
            direction = "Left"
            move_size = 0.3
            x -= move_size
        elif rn < 80:  # 15% Right
            direction = "Right"
            move_size = 0.4
            x += move_size
        elif rn < 90:  # 10% Up
            direction = "Up"
            move_size = 0.2
            z += move_size
        else:          # 10% Down
            direction = "Down"
            move_size = 0.3
            z -= move_size
            
        # Display the movement and current coordinates
        print(f"{step:<5} {rn:<4} {direction:<12} {move_size:<10} ({x:>6.1f}, {y:>6.1f}, {z:>6.1f})")

    print("-" * 65)
    print(f"Final 3D Position after {steps} steps: (x: {x:.2f}, y: {y:.2f}, z: {z:.2f})")

# Run for 50 steps
simulate_3d_random_walk(50)
```
