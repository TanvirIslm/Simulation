```python
import numpy as np
import matplotlib.pyplot as plt

# Bernoulli parameter
p = 0.85  # Probability of success

# Define X
print("Bernoulli Distribution - Quality Control Test")
print("---------------------------------------------")
print("X = 1 : Chip passes the test (Success)")
print("X = 0 : Chip fails the test (Failure)")

# Probability of defect
prob_failure = 1 - p

# Mean and Variance
mean = p
variance = p * (1 - p)

print("\nResults")
print("P(X = 0) =", round(prob_failure, 4))
print("E[X] =", round(mean, 4))
print("Var(X) =", round(variance, 4))

# Simulation
n = 1000
samples = np.random.binomial(1, p, n)

successes = np.sum(samples)
failures = n - successes

print("\nSimulation Results")
print("Total Chips Tested =", n)
print("Passed =", successes)
print("Failed =", failures)

# Plot
plt.figure(figsize=(6,4))
plt.bar(["Failure (0)", "Success (1)"],
        [failures, successes])

plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.title("Bernoulli Distribution of Chip Testing")
plt.grid(axis='y')
plt.show()
```
