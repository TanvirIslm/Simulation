```python
import numpy as np
from scipy import stats

# -----------------------------------
# Simulation Parameters
# -----------------------------------

NUM_REPLICATIONS = 30
NUM_CUSTOMERS = 100

MEAN_INTERARRIVAL = 2.0
MEAN_SERVICE = 1.5

average_waiting_times = []

# -----------------------------------
# Run Independent Replications
# -----------------------------------

for rep in range(NUM_REPLICATIONS):

    arrival_time = 0
    server_free_time = 0

    total_wait = 0

    for customer in range(NUM_CUSTOMERS):

        interarrival = np.random.exponential(
            MEAN_INTERARRIVAL
        )

        service_time = np.random.exponential(
            MEAN_SERVICE
        )

        arrival_time += interarrival

        wait_time = max(
            0,
            server_free_time - arrival_time
        )

        total_wait += wait_time

        service_start = max(
            arrival_time,
            server_free_time
        )

        server_free_time = (
            service_start + service_time
        )

    avg_wait = total_wait / NUM_CUSTOMERS

    average_waiting_times.append(avg_wait)

# -----------------------------------
# Confidence Interval Calculation
# -----------------------------------

sample_mean = np.mean(
    average_waiting_times
)

sample_std = np.std(
    average_waiting_times,
    ddof=1
)

t_value = stats.t.ppf(
    0.975,
    df=NUM_REPLICATIONS - 1
)

margin_error = (
    t_value
    * sample_std
    / np.sqrt(NUM_REPLICATIONS)
)

lower = sample_mean - margin_error
upper = sample_mean + margin_error

# -----------------------------------
# Results
# -----------------------------------

print("CONFIDENCE INTERVAL ESTIMATION")
print("-" * 40)

print("Number of Replications:",
      NUM_REPLICATIONS)

print("Average Waiting Times:")
print(np.round(
    average_waiting_times,
    4
))

print("\nSample Mean:",
      round(sample_mean, 4))

print("Sample Standard Deviation:",
      round(sample_std, 4))

print("\n95% Confidence Interval:")
print(
    "(",
    round(lower, 4),
    ",",
    round(upper, 4),
    ")"
)
```
