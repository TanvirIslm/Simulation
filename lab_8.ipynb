```python
import numpy as np
import matplotlib.pyplot as plt
from math import exp

# ----------------------------------
# Part 1: COVID Wave Problem
# ----------------------------------

mean_days = 100

# Rate parameter λ = 1/mean
lam = 1 / mean_days

t = 120

# P(X > t) = e^(-λt)
probability = exp(-lam * t)

print("Exponential Distribution")
print("------------------------")
print("Average interval between waves =", mean_days, "days")
print("Rate parameter (λ) =", round(lam, 4))
print("P(X > 120) =", round(probability, 4))

# ----------------------------------
# Part 2: Simulate Exponential Distributions
# ----------------------------------

rates = [0.5, 1.0, 2.0, 4.0]

plt.figure(figsize=(10, 6))

for rate in rates:

    sample = np.random.exponential(
        scale=1/rate,
        size=1000
    )

    plt.hist(
        sample,
        bins=40,
        density=True,
        alpha=0.5,
        label=f'λ = {rate}'
    )

plt.xlabel("Time")
plt.ylabel("Density")
plt.title("Exponential Distributions for Different Rates")
plt.legend()
plt.grid(True)
plt.show()

# ----------------------------------
# Part 3: PDF Curves
# ----------------------------------

x = np.linspace(0, 10, 1000)

plt.figure(figsize=(10, 6))

for rate in rates:

    y = rate * np.exp(-rate * x)

    plt.plot(
        x,
        y,
        linewidth=2,
        label=f'λ = {rate}'
    )

plt.xlabel("Time")
plt.ylabel("f(x)")
plt.title("Exponential Distribution PDF")
plt.legend()
plt.grid(True)
plt.show()
```
