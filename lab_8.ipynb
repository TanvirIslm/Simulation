```python
import numpy as np
import matplotlib.pyplot as plt
from math import exp

# ----------------------------------
# Part 1: COVID Wave Problem
# ----------------------------------

mean_days = 100

# Rate parameter λ = 1/mean
lam = 1 / mean_days

t = 120

# P(X > t) = e^(-λt)
probability = exp(-lam * t)

print("Exponential Distribution")
print("------------------------")
print("Average interval between waves =", mean_days, "days")
print("Rate parameter (λ) =", round(lam, 4))
print("P(X > 120) =", round(probability, 4))

# ----------------------------------
# Part 2: Simulate Exponential Distributions
# ----------------------------------

rates = [0.5, 1.0, 2.0, 4.0]

plt.figure(figsize=(10, 6))

for rate in rates:

    sample = np.random.exponential(
        scale=1/rate,
        size=1000
    )

    plt.hist(
        sample,
        bins=40,
        density=True,
        alpha=0.5,
        label=f'λ = {rate}'
    )

plt.xlabel("Time")
plt.ylabel("Density")
plt.title("Exponential Distributions for Different Rates")
plt.legend()
plt.grid(True)
plt.show()

# ----------------------------------
# Part 3: PDF Curve for COVID Example
# ----------------------------------

x = np.linspace(0, 200, 1000)

# PDF
y = lam * np.exp(-lam * x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, color='blue', linewidth=2, label='Exponential PDF')

# Mark 120-3, 120, 120+3
points = [117, 120, 123]

for p in points:
    py = lam * np.exp(-lam * p)
    plt.scatter(p, py, color='red')
    plt.vlines(p, 0, py, colors='red', linestyles='dashed')
    plt.text(p + 1, py, f'({p}, {py:.4f})')

plt.xlabel("Time (Days)")
plt.ylabel("Probability Density")
plt.title("Exponential Distribution (Mean = 100 Days)")
plt.legend()
plt.grid(True)
plt.show()
```
