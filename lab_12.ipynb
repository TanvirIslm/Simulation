```python
import random
import math
import matplotlib.pyplot as plt

# Parameters
lam = 0.5
n = 1000

# Generate Random Variates
random_variates = []

for i in range(n):

    u = random.random()

    x = -(1 / lam) * math.log(1 - u)

    random_variates.append(x)

# Print first 20 values
print("First 20 Random Variates:\n")

for i in range(20):
    print(round(random_variates[i], 4))

# Histogram
plt.figure(figsize=(8,5))
plt.hist(
    random_variates,
    bins=30,
    density=True
)

plt.xlabel("Random Variate")
plt.ylabel("Density")
plt.title("Inverse Transform Method (Exponential Distribution)")
plt.grid(True)

plt.show()
```
