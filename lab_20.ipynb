```python
import numpy as np
from scipy.stats import t
import matplotlib.pyplot as plt
import math

# ---------------------------------------
# Average Waiting Times (5 Replications)
# ---------------------------------------
data = [3.2, 4.3, 5.1, 4.2, 4.6]

n = len(data)

# ---------------------------------------
# Statistical Calculations
# ---------------------------------------

# Sample Mean
mean = np.mean(data)

# Sample Variance
variance = np.var(data, ddof=1)

# Sample Standard Deviation
std = np.std(data, ddof=1)

# Standard Error
se = std / math.sqrt(n)

# t-value for 95% Confidence Interval
t_value = t.ppf(0.975, n - 1)

# Margin of Error
margin = t_value * se

# Confidence Interval
lower = mean - margin
upper = mean + margin

# Desired Precision
h = 0.5

# Required Number of Replications
required_replications = math.ceil((t_value * std / h) ** 2)

# ---------------------------------------
# Results
# ---------------------------------------

print("========== OUTPUT ANALYSIS ==========")

print("Sample Mean =", round(mean,3))

print("Sample Variance =", round(variance,3))

print("Standard Deviation =", round(std,3))

print("Standard Error =", round(se,3))

print("95% Confidence Interval = ({:.3f}, {:.3f})".format(lower, upper))

print("Required Replications =", required_replications)

# =======================================
# Plot 1 : Bar Chart
# =======================================

plt.figure(figsize=(8,5))

plt.bar(range(1, n+1), data)

plt.title("Average Waiting Time per Replication")

plt.xlabel("Replication")

plt.ylabel("Average Waiting Time (minutes)")

plt.xticks(range(1, n+1))

plt.grid(axis='y')

plt.show()

# =======================================
# Plot 2 : Mean with 95% Confidence Interval
# =======================================

plt.figure(figsize=(6,5))

plt.errorbar(
    1,
    mean,
    yerr=margin,
    fmt='o',
    capsize=8,
)

plt.xlim(0.5,1.5)

plt.xticks([1],["Mean"])

plt.title("95% Confidence Interval")

plt.ylabel("Waiting Time (minutes)")

plt.grid(True)

plt.show()

# =======================================
# Plot 3 : Histogram
# =======================================

plt.figure(figsize=(8,5))

plt.hist(
    data,
    bins=5,
    edgecolor='black'
)

plt.title("Distribution of Average Waiting Times")

plt.xlabel("Average Waiting Time (minutes)")

plt.ylabel("Frequency")

plt.grid(True)

plt.show()
```
