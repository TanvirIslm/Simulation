```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import exp, factorial

# -----------------------------
# Part 1: λ = 5 calls/hour
# -----------------------------
lam = 5

x_values = list(range(11))
probabilities = []

for x in x_values:
    p = (exp(-lam) * (lam ** x)) / factorial(x)
    probabilities.append(p)

# Table
df = pd.DataFrame({
    "Number of Calls (x)": x_values,
    "P(X=x)": probabilities
})

print("Poisson Distribution (λ = 5)\n")
print(df.round(6))

# PMF Graph for λ = 5
plt.figure(figsize=(8,5))
plt.bar(x_values, probabilities)
plt.xlabel("Number of Calls")
plt.ylabel("Probability")
plt.title("Poisson PMF (λ = 5)")
plt.grid(True)
plt.show()

# -----------------------------
# Part 2: Compare λ=10 and λ=15
# -----------------------------
x = np.arange(0, 31)

pmf_10 = [(exp(-10) * (10**k)) / factorial(k) for k in x]
pmf_15 = [(exp(-15) * (15**k)) / factorial(k) for k in x]

plt.figure(figsize=(10,5))

plt.plot(x, pmf_10, marker='o', label='λ = 10')
plt.plot(x, pmf_15, marker='s', label='λ = 15')

plt.xlabel("Number of Calls")
plt.ylabel("Probability")
plt.title("Poisson PMF Comparison")
plt.legend()
plt.grid(True)

plt.show()
```
