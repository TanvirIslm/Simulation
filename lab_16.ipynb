```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------------
# Parameters
# -----------------------------------

SIM_TIME = 1000
WARM_UP = 200

production_times = []

# -----------------------------------
# Continuous Manufacturing Simulation
# -----------------------------------

for t in range(SIM_TIME):

    # Processing time follows exponential distribution
    process_time = np.random.exponential(scale=5)

    production_times.append(process_time)

# -----------------------------------
# Warm-up Analysis
# -----------------------------------

running_mean = []

for i in range(1, len(production_times) + 1):
    running_mean.append(
        np.mean(production_times[:i])
    )

# -----------------------------------
# Steady-State Mean
# -----------------------------------

steady_state_data = production_times[WARM_UP:]

steady_state_mean = np.mean(
    steady_state_data
)

# -----------------------------------
# Results
# -----------------------------------

print("STEADY-STATE ANALYSIS")
print("-" * 40)

print("Simulation Length =", SIM_TIME)
print("Warm-up Period =", WARM_UP)

print("\nSteady-State Mean Time =",
      round(steady_state_mean, 4))

# -----------------------------------
# Running Mean Plot
# -----------------------------------

plt.figure(figsize=(10,5))

plt.plot(
    running_mean,
    label="Running Mean"
)

plt.axvline(
    WARM_UP,
    linestyle='--',
    label="Warm-up End"
)

plt.xlabel("Observation")
plt.ylabel("Mean Processing Time")
plt.title("Warm-up Period Analysis")
plt.legend()
plt.grid(True)

plt.show()

# -----------------------------------
# Histogram of Steady-State Data
# -----------------------------------

plt.figure(figsize=(8,5))

plt.hist(
    steady_state_data,
    bins=30,
    density=True
)

plt.xlabel("Processing Time")
plt.ylabel("Density")
plt.title("Steady-State Distribution")
plt.grid(True)

plt.show()
```
