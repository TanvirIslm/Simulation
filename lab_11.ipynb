```python
import matplotlib.pyplot as plt

# LCG Parameters
a = 1664525
c = 1013904223
m = 2**32
seed = 12345

n = 100

# Generate Random Numbers
x = seed
random_numbers = []

for i in range(n):
    x = (a * x + c) % m
    random_numbers.append(x / m)

# Print Numbers
print("First 20 Random Numbers:")
for i in range(20):
    print(round(random_numbers[i], 6))

# Plot Sequence
plt.figure(figsize=(8,5))
plt.plot(range(n), random_numbers, marker='o')

plt.xlabel("Iteration")
plt.ylabel("Random Number")
plt.title("Linear Congruential Generator (LCG)")
plt.grid(True)
plt.show()

# Histogram
plt.figure(figsize=(8,5))
plt.hist(random_numbers, bins=10)

plt.xlabel("Random Number")
plt.ylabel("Frequency")
plt.title("Histogram of LCG Random Numbers")
plt.grid(True)
plt.show()
```
