```python
from math import comb
import matplotlib.pyplot as plt

# Parameters
n = 15
p = 0.90
q = 1 - p

# Example Data
data = [14, 13, 15, 12, 14]

# -----------------------------
# 1. Exact Probability P(X=13)
# -----------------------------
x = 13
probability = comb(n, x) * (p ** x) * (q ** (n - x))

# -----------------------------
# 2. Theoretical Mean & Variance
# -----------------------------
mean = n * p
variance = n * p * q

# -----------------------------
# 3. Sample Mean
# -----------------------------
sample_mean = sum(data) / len(data)

# -----------------------------
# Results
# -----------------------------
print("NETWORK PACKET DELIVERY")
print("-" * 40)

print("Parameters:")
print("n =", n)
print("p =", p)
print("q =", q)

print("\nProbability of exactly 13 successful packets:")
print("P(X=13) =", round(probability, 6))

print("\nTheoretical Mean (μ):")
print("μ =", mean)

print("\nTheoretical Variance (σ²):")
print("σ² =", variance)

print("\nExample Data:")
print(data)

print("\nSample Mean:")
print(round(sample_mean, 2))

print("\nDifference from Theoretical Mean:")
print(round(abs(sample_mean - mean), 2))

# -----------------------------
# Visualization
# -----------------------------
batch_numbers = list(range(1, len(data) + 1))

plt.figure(figsize=(8, 5))
plt.bar(batch_numbers, data)

plt.axhline(mean, linestyle='--',
            label=f"Theoretical Mean = {mean}")

plt.xlabel("Batch Number")
plt.ylabel("Successful Packets")
plt.title("Packet Delivery per Batch")
plt.legend()
plt.grid(True)

plt.show()
```
